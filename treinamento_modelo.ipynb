```python
!pip install ultralytics roboflow

from roboflow import Roboflow
rf = Roboflow(api_key="chave")
project = rf.workspace("").project("")
version = project.version(2)
dataset = version.download("yolov8")

import yaml
yaml_path = f"{dataset.location}/data.yaml"

with open(yaml_path, 'r') as f:
    dados_yaml = yaml.safe_load(f)

dados_yaml['train'] = f"{dataset.location}/train/images"
dados_yaml['val'] = f"{dataset.location}/train/images"

with open(yaml_path, 'w') as f:
    yaml.dump(dados_yaml, f)

print("\n Novo Dataset baixado, corrigido e roteado com sucesso!\n")

from ultralytics import YOLO

model = YOLO("visdrone.pt")
resultados = model.train(data=yaml_path, epochs=100, imgsz=640, device=0)
```
